```python
random.shuffle(passwords_tuple) #shuffling randomly for robustness
```

```python
x=[label[0] for label in passwords_tuple]
```

```python
x
```

```python
sns.countplot(x="strength",data=data)
plt.show()
```

```python
data.shape
```

```python
def word_devide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_devide_char)
x=vectorizer.fit_transform(x)
```

```python
x.shape
```

```python
vectorizer.vocabulary_
```

```python
data.iloc[0]
```

```python
feature_names=vectorizer.get_feature_names()
```

```python
df=pd.DataFrame(x[0].T.todense(),index=feature_names,columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x.shape
```

```python
len(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.20)
```

```python
log_classifier=LogisticRegression(penalty="l2",multi_class="ovr")
```

```python
log_classifier.fit(x_train,y_train)
```

```python
print(log_classifier.score(x_test,y_test))
```

```python
# mutinomial
log_clf=LogisticRegression(penalty="l2",multi_class="multinomial",solver="newton-cg")
log_clf.fit(x_train,y_train)
```

```python
log_clf.score(x_test,y_test)
```

```python
pwd=np.array(["mufsi@123"])
pwd=vectorizer.transform(pwd)
log_clf.predict(pwd)
```

## XGBoost

```python
from xgboost import XGBClassifier
```

```python
xgbclassifier=XGBClassifier()
xgbclassifier.fit(x_train,y_train)
```

```python
xgbclassifier.score(x_test,y_test)
```

```python
pwd=np.array(["mufseera@12"])
pwd=vectorizer.transform(pwd)
xgbclassifier.predict(pwd)
```
